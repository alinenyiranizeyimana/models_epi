$$\textbf{Lagrange Intepolation}$$

$$ P(x) = \sum_{i = 0} ^ n L_i(x) y_i$$

$$..............................$$
$$ L_i(x) = \prod_{j = 0} ^n \frac{(x- x_j)}{x_i - x_j}$$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,120, 20)
x= np.array(x)


""" Y computed using linear regression
n degree of polynomial
"""

m=len(x)
n = m-1

x_poly = 0
y = np.array([26.0, 48.6, 61.6, 71.2, 74.8,75.2],float)
lagrange_list =[]


x_plot = np.linspace(0,120,30)
y_plot = []

for x_poly in x_plot:
    y_value = 0
    for i in range(n+1):
        lagrange =1
        for j in range(n+1):
            if (j != i):
                lagrange *= (x_poly - x[j])/(x[i] - x[j])
                      
        lagrange_list.append(lagrange)
        y_value += lagrange * y[i]

    y_plot.append(y_value)

   
"""
print("Lagrange value", lagrange_list)
print("Same Y:",y_value )
"""
plt.plot(x_plot, y_plot, marker="*", ms =12, mfc= "red", mec ="white")
plt.show()
    
```
